## a)  A continuación, se presenta una transformación sobre ambas imágenes 

<br> <center> $ G(x,y) = (1-k)*I1(x,y) + k*I2(x,y) $ </center> </br>

<br> <center> $ 0 < k < 1 $ </center> </br>

<ul> Realice un código que muestre esta transformación, posteriormente responda: </ul>

<ul> <br> ¿Qué ocurre si k = 1, k = 0, k = 0.5? </br> </ul>

```python
#Se multiplican los valores de cada canal de las imagenes por (1-K) unidades, con 0 < K < 1
#Para k =1

k = 0.5
p = (1-k)

#Se crea la imagen copia
img1_div = np.zeros(img1.shape, np.uint8)
img2_div = np.zeros(img2.shape, np.uint8)

#Se sustituyen los valores de cada canal de las copias por los valores modificados de la imagen original

#se multiplca por (1-k) la imagen 1
img1_div[:,:,0] = cv2.multiply(img1[:,:,0], p)
img1_div[:,:,1] = cv2.multiply(img1[:,:,1], p)
img1_div[:,:,2] = cv2.multiply(img1[:,:,2], p)


#se multiplca por k la imagen 2
img2_div[:,:,0] = cv2.multiply(img2[:,:,0], k)
img2_div[:,:,1] = cv2.multiply(img2[:,:,1], k)
img2_div[:,:,2] = cv2.multiply(img2[:,:,2], k)

#hacerlas del mismo tamaño para poderlas sumar

img2 = cv2.resize(img2, (img1.shape[1], img1.shape[0]))

#Indicamos el parámetro a que indica la importancia de cada imagen


#Mediante la T.Lineal, si a=1, tenemos la imagen 1 únicamente. Si a=0, obtenemos solo img2.
# Si 0<a<1 tendremos combinaciones ponderadas de las imágenes entrada
res = cv2.add(img2*k,img1*(1-k)).astype(np.uint8)


#Se realiza el arreglo de 2x2 para mostrar imagen modificada vs histogramas
fig, arreglo_plots = plt.subplots(1,4, figsize=(15, 3))
colors = ('r','g','b')

#Se le dan títulos a los cuadros, se imprimen la img 1 y su histograma
arreglo_plots[0].set_title('Imagen original')
arreglo_plots[0].imshow(img1, cmap="gray")
arreglo_plots[1].set_title('Histograma original')
for i,col in enumerate(colors):
    img_array_i = img1[:,:,i].ravel()
    arreglo_plots[1].hist(img_array_i,histtype='step', bins=255, 
                            range=(0.0, 255.0),normed=True, color=colors[i])

arreglo_plots[2].set_title('Imagen modificada k=1')
arreglo_plots[2].imshow(img1_div, cmap="gray")
arreglo_plots[3].set_title('Histograma modificado k=1')
for i,col in enumerate(colors):
    img_array_i = img1_div[:,:,i].ravel()
    arreglo_plots[3].hist(img_array_i,histtype='step', bins=255, 
                            range=(0.0, 255.0),normed=True, color=colors[i])
    
fig,arreglo_plots2 = plt.subplots(1,4, figsize=(15, 3))
    
#Se le dan títulos a los cuadros, se imprimen la img2 y su histograma
arreglo_plots2[0].set_title('Imagen original')
arreglo_plots2[0].imshow(img2, cmap="gray")
arreglo_plots2[1].set_title('Histograma original')
for i,col in enumerate(colors):
    img_array_i = img2[:,:,i].ravel()
    arreglo_plots2[1].hist(img_array_i,histtype='step', bins=255, 
                            range=(0.0, 255.0),normed=True, color=colors[i])

arreglo_plots2[2].set_title('Imagen modificada k=1')
arreglo_plots2[2].imshow(img2_div, cmap="gray")
arreglo_plots2[3].set_title('Histograma modificado k=1')
for i,col in enumerate(colors):
    img_array_i = img2_div[:,:,i].ravel()
    arreglo_plots2[3].hist(img_array_i,histtype='step', bins=255, 
                            range=(0.0, 255.0),normed=True, color=colors[i])    
    
fig,arreglo_plots3 = plt.subplots(1,2, figsize=(15, 3))

#Se le dan títulos a los cuadros, se imprimen la suma y su histograma
arreglo_plots3[0].set_title('Imagen suma')
arreglo_plots3[0].imshow(res, cmap="gray")
arreglo_plots3[1].set_title('Histograma suma')
for i,col in enumerate(colors):
    img_array_i = res[:,:,i].ravel()
    arreglo_plots3[1].hist(img_array_i,histtype='step', bins=255, 
                            range=(0.0, 255.0),normed=True, color=colors[i])    
```
